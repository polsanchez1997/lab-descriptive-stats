```python
orders['InvoiceDate'] = pd.to_datetime(orders['InvoiceDate'])

orders['Year'] = orders['InvoiceDate'].dt.year

average_amount_per_year = orders.groupby('Year')['amount_spent'].mean().reset_index()

average_amount_per_year
```
